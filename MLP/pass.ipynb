```python
from find import *
import pandas as pd
```

```python
# Remove unneeded columns
df = pd.read_csv('NFL TRACKING/2017091100.csv')
df = df.drop(columns=['nflId', 'displayName', 'frameId', 'gameId', 'time']).rename(columns={'Unnamed: 0': 'index'})
df.to_csv('cleaned.csv', index=False)
```

```python
# Find all passes
searchCSV('cleaned.csv', [('event', '.pass')], 'passes.csv')
```

```python
# Get all plays that involves passing
df_passes = pd.read_csv('passes.csv')
playIds = df_passes['playId'].unique()
```

```python
df_plays = pd.DataFrame()
for playId in playIds:
    result = searchCSV('cleaned.csv', [('playId', str(playId))])
    df_plays = pd.concat([df_plays, result], ignore_index=True)

df_plays.head(3)
```

```python
df_plays.to_csv('temp.csv', index=False)
```

```python
df_pass_forwards = searchDF(df_plays, [('event', 'pass_forward')])
df_pass_forwards.head(22)
```

```python
df_pass_forwards['result'] = None
# pass_arrived pass_outcome_caught, pass_outcome_incomplete, pass_outcome_interception, pass_outcome_touchdown, pass_outcome_tipped
for playId in playIds:
    events = searchDF(df_passes, [('playId', str(playId))])['event'].values
    if 'pass_outcome_caught' in events:
        result = 1
    elif 'pass_outcome_incomplete' in events:
        result = 2
    elif 'pass_outcome_interception' in events:
        result = 3
    elif 'pass_outcome_touchdown' in events:
        result = 4
    elif 'pass_outcome_tipped' in events:
        result = 5
    elif 'pass_arrived' in events:
        result = 0
    else:
        raise Exception(f'Events does not contain any passes in {playId}')
    df_pass_forwards.loc[df_pass_forwards['playId'] == str(playId), 'result'] = result

df_pass_forwards = df_pass_forwards.drop('event', axis=1)
df_pass_forwards['team'] = df_pass_forwards['team'].replace({'home':0, 'away':3})
df_pass_forwards.rename(columns={'team':'role'}, inplace=True)
df_pass_forwards.head(44)
```

```python
df_pass_forwards.to_csv('data.csv', index=False)
```

```python
class Node:
    def __init__(self, jerseyNumber, x, y, speed, distance, direction, role):
        self.number = jerseyNumber # Not used in prediction
        self.x = x
        self.y = y
        self.speed = speed
        self.distance = distance
        self.direction = direction
        self.role = role # 0 = home, 1 = initiator, 2 = receiver, 3 = away
```

```python
df = pd.read_csv('data.csv')
playId = 152
team = 3 # 0 = home(blue), 3 = away(white)
init = 9
recv = 41

# viking home, saints away

# Initiator
print(f"Initiator exist: {((df['playId'] == playId) & (df['jerseyNumber'] == float(init)) & (df['role'] == team)).any()}")
df.loc[(df['playId'] == playId) & (df['jerseyNumber'] == float(init)) & (df['role'] == team), 'role'] = 1
# Receiver
print(f"Receiver exist: {((df['playId'] == playId) & (df['jerseyNumber'] == float(recv)) & (df['role'] == team)).any()}")
df.loc[(df['playId'] == playId) & (df['jerseyNumber'] == float(recv)) & (df['role'] == team), 'role'] = 2

df.to_csv('data.csv', index=False)
```
